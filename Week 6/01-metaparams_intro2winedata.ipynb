```python
### import libraries

# for DL modeling
import torch
from torch.utils.data import DataLoader,TensorDataset
from sklearn.model_selection import train_test_split

# for number-crunching
import numpy as np
import scipy.stats as stats

# for dataset management
import pandas as pd

# for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import and process the data

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

data = pd.read_csv(url,sep=';')
data
```

```python
# describe the data
data.describe()
```

```python
# list number of unique values per column

for i in data.keys():
  print(f'{i} has {len(np.unique(data[i]))} unique values')
```

```python
# Import the necessary libraries
import seaborn as sns  # Import the seaborn library for data visualization
import matplotlib.pyplot as plt  # Import the matplotlib.pyplot library for displaying plots

# Define the columns to be plotted in the pairwise plot
cols2plot = ['fixed acidity', 'volatile acidity', 'citric acid', 'quality']
# Create a list called 'cols2plot' that contains the names of the columns to be included in the pairwise plot.

# Create a pairwise scatterplot matrix with regression lines and colored by 'quality'
sns.pairplot(data[cols2plot], kind='reg', hue='quality')
# Use the 'sns.pairplot()' function to create a scatterplot matrix of the selected columns.
# - 'data[cols2plot]' specifies the data to be used for the plot.
# - 'kind='reg'' specifies that regression lines should be added to the scatterplots.
# - 'hue='quality'' colors the points in the scatterplot matrix based on the 'quality' column in the dataset.

# Display the plot
plt.show()
# Use 'plt.show()' to display the generated plot.
```

```python
# plot some data
fig,ax = plt.subplots(1,figsize=(17,4))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.show()


# remove rows with outliers
# data = data[data['total sulfur dioxide']<200]
```

```python
### z-score all variables except for quality

# Find the columns we want to normalize (all except quality)
cols2zscore = data.keys()  # Get the column names from the 'data' DataFrame
cols2zscore = cols2zscore.drop('quality')  # Remove the 'quality' column from the list of columns to be normalized

# Z-score normalization (written out for clarity)
for col in cols2zscore:
  meanval   = np.mean(data[col])  # Calculate the mean (average) value of the current column
  stdev     = np.std(data[col], ddof=1)  # Calculate the standard deviation of the current column with Bessel's correction (ddof=1)
  data[col] = (data[col] - meanval) / stdev  # Apply the Z-score formula to standardize the values in the current column

# You can also perform Z-score normalization more compactly using the 'apply' method
# data[cols2zscore] = data[cols2zscore].apply(stats.zscore)

# Display summary statistics of the DataFrame after normalization
data.describe()
```

```python
# check the plot again
fig,ax = plt.subplots(1,figsize=(17,4))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.show()
```

```python
# distribution quality values
fig = plt.figure(figsize=(10,7))
plt.rcParams.update({'font.size': 22}) # increase font size in the figure

counts = data['quality'].value_counts()
plt.bar(list(counts.keys()),counts)
plt.xlabel('Quality rating')
plt.ylabel('Count')
plt.show()

# create a new column for binarized (boolean) quality
data['boolQuality'] = 0
# data['boolQuality'][data['quality']<6] = 0 # implicit in the code! just here for clarity
data['boolQuality'][data['quality']>5] = 1

data[['quality','boolQuality']]
```

# Re-organize the data: train/test in DataLoaders

```python
# Convert from pandas DataFrame to tensor
dataT  = torch.tensor(data[cols2zscore].values).float()
# Create a tensor 'dataT' containing the values from selected columns in the 'data' DataFrame.
# - 'data[cols2zscore].values' extracts the values from the columns specified in 'cols2zscore'.
# - 'torch.tensor()' converts the values into a PyTorch tensor.
# - '.float()' ensures that the tensor contains floating-point values.

labels = torch.tensor(data['boolQuality'].values).float()
# Create a tensor 'labels' containing the values from the 'boolQuality' column in the 'data' DataFrame.
# - 'data['boolQuality'].values' extracts the values from the 'boolQuality' column.
# - 'torch.tensor()' converts the values into a PyTorch tensor.
# - '.float()' ensures that the tensor contains floating-point values.

print(dataT.shape)
# Print the shape (dimensions) of the 'dataT' tensor.
# This provides information about the number of rows and columns in the tensor.

print(labels.shape)
# Print the shape (dimensions) of the 'labels' tensor.
# This provides information about the number of elements in the tensor.
# 
# We'll actually need the labels to be a "tensor"
labels = labels[:, None]
# Reshape the 'labels' tensor by adding an additional dimension.
# This effectively converts a 1D tensor into a 2D tensor with a single column.
# The resulting shape is (number of samples, 1), which is often needed for compatibility with certain machine learning algorithms.
# The 'None' indexing adds a new dimension along the second axis (columns).
```

```python
# Use scikit-learn to split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(dataT, labels, test_size=.1)
# Split the data into training and testing sets using scikit-learn's 'train_test_split' function.
# - 'dataT': The feature data stored in the 'dataT' tensor.
# - 'labels': The corresponding labels stored in the 'labels' tensor.
# - 'test_size=.1': Specifies that 10% of the data should be allocated for testing, and the remaining 90% for training.

# Then convert them into PyTorch Datasets (note: already converted to tensors)
train_data = TensorDataset(train_data, train_labels)
test_data  = TensorDataset(test_data, test_labels)
# Convert the training and testing data along with their labels into PyTorch 'TensorDataset' objects.
# 'TensorDataset' is a PyTorch dataset wrapper that combines feature data and labels into a single dataset.

# Finally, translate into DataLoader objects
batchsize    = 64
train_loader = DataLoader(train_data, batch_size=batchsize, shuffle=True,drop_last=True)
test_loader  = DataLoader(test_data, batch_size=test_data.tensors[0].shape[0])
# Create DataLoader objects for both the training and testing datasets.
# - 'train_data' and 'test_data' are the 'TensorDataset' objects containing feature data and labels.
# - 'batch_size': Specifies the number of samples in each mini-batch. Here, it's set to 64 for the training loader.
# - 'shuffle=True': Shuffles the training data to introduce randomness during training.
# - 'test_data.tensors[0].shape[0]': Sets the batch size of the testing loader to the total number of samples in the testing dataset.
```
