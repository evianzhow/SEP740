```python
# import libraries
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader,TensorDataset
from sklearn.model_selection import train_test_split

# summary info on models
from torchsummary import summary

import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

# Create Gaussian blurs with different widths

- **Synthetic Gaussian Image Generation**: This segment of code illustrates the process of creating a dataset consisting of synthetic images, each representing a Gaussian blob. The purpose is to generate a balanced dataset with two classes of Gaussian blobs, distinguished by their widths. This dataset can serve as a simple yet effective testbed for training and evaluating machine learning models, particularly in tasks requiring the classification of visual patterns.

- **Parameter Setup and Grid Definition**: A set number of samples per class (`nPerClass`) and the size of the images (`imgSize`) are defined. A spatial grid spanning from -4 to 4 across both dimensions is created using `np.meshgrid`, which serves as the foundation for generating Gaussian distributions.

- **Gaussian Blob Specifications**: Two different standard deviations (`widths`) are specified for the Gaussian blobs, resulting in blobs of different visual widths. These serve as the basis for the two classes within the dataset.

- **Image and Label Tensor Initialization**: Tensors `images` and `labels` are initialized to store the generated images and their corresponding class labels. Each image is stored as a single-channel grayscale image of size `imgSize x imgSize`.

- **Gaussian Blob Creation and Noise Addition**: For each sample, a Gaussian blob is generated with a random center within the defined spatial grid. The width of the Gaussian is determined by cycling between the specified `widths`, thereby assigning blobs to one of two classes. Gaussian noise is then added to each image to introduce variability and mimic real-world imperfections.

- **Tensor Population**: The generated Gaussian images and their labels are added to the `images` and `labels` tensors, respectively. Each label corresponds to the class of the Gaussian blob, determined by the width used in its generation.

- **Label Tensor Reshaping**: Finally, the `labels` tensor is reshaped to ensure its dimensions align with typical expectations for classification tasks in PyTorch, where labels are often expected to be in a specific shape for compatibility with loss functions and model outputs.

This markdown content describes the methodology behind generating a synthetic dataset of Gaussian blobs with varying widths, emphasizing the dataset's structure, the generation process, and the intended use case for machine learning applications.


```python
nPerClass = 1000
imgSize   = 91

x = np.linspace(-4,4,imgSize)
X,Y = np.meshgrid(x,x)

# the two widths (a.u.)
widths = [1.8,2.4]

# initialize tensors containing images and labels
images = torch.zeros(2*nPerClass,1,imgSize,imgSize)
labels = torch.zeros(2*nPerClass)

for i in range(2*nPerClass):

  # create the gaussian with random centers
  ro = 2*np.random.randn(2) # ro = random offset
  G  = np.exp( -( (X-ro[0])**2 + (Y-ro[1])**2) / (2*widths[i%2]**2) )
  
  # and add noise
  G  = G + np.random.randn(imgSize,imgSize)/5
  
  # add to the tensor
  images[i,:,:,:] = torch.Tensor(G).view(1,imgSize,imgSize)
  labels[i] = i%2

labels = labels[:,None]
```

```python
# visualize some images
fig,axs = plt.subplots(3,7,figsize=(13,6))

for i,ax in enumerate(axs.flatten()):
  whichpic = np.random.randint(2*nPerClass)
  G = np.squeeze( images[whichpic,:,:] )
  ax.imshow(G,vmin=-1,vmax=1,cmap='jet')
  ax.set_title('Class %s'%int(labels[whichpic].item()))
  ax.set_xticks([])
  ax.set_yticks([])

plt.show()
```

# Create train/test groups using DataLoader

```python
# Step 2: use scikitlearn to split the data
train_data,test_data, train_labels,test_labels = train_test_split(images, labels, test_size=.1)

# Step 3: convert into PyTorch Datasets
train_data = TensorDataset(train_data,train_labels)
test_data  = TensorDataset(test_data,test_labels)

# Step 4: translate into dataloader objects
batchsize    = 32
train_loader = DataLoader(train_data,batch_size=batchsize,shuffle=True,drop_last=True)
test_loader  = DataLoader(test_data,batch_size=test_data.tensors[0].shape[0])
```

```python
# check size (should be images X channels X width X height
print( train_loader.dataset.tensors[0].shape )
print( train_loader.dataset.tensors[1].shape )
```

# Create the DL model

- **Gaussian Blob Classification Model**: This code outlines the creation of a convolutional neural network (CNN) called `gausnet`, designed for the classification of synthetic Gaussian blob images. The network architecture is defined using `nn.Module`, with layers organized sequentially using `nn.Sequential`.

- **Sequential Layer Definition**: The `gausnet` model comprises a series of convolutional layers, ReLU activation functions, and average pooling layers, structured to progressively reduce the spatial dimensions of the input images while extracting relevant features. The sequence begins with a convolutional layer that maintains the input size through padding, followed by ReLU activation and average pooling to reduce dimensions. This pattern repeats, further condensing the feature maps, before the data is flattened for processing by fully connected layers.

- **Fully Connected Layers**: After the convolutional and pooling layers, the network transitions to fully connected layers. The first linear layer transforms the flattened convolutional output into a 50-dimensional space, followed by a final linear layer that produces a single output. This output is intended to represent the model's confidence in classifying the input image as one of two Gaussian blob types.

- **Forward Pass Functionality**: The `forward` method applies the defined sequence of operations to input data `x`, facilitating the model's ability to make predictions based on learned features.

- **Model Instantiation and Training Setup**: Upon defining the `gausnet` class, an instance of the model is created, and a binary cross-entropy loss function with logits (`BCEWithLogitsLoss`) is specified for training. This choice of loss function is appropriate for binary classification tasks. An Adam optimizer is used with a learning rate of 0.001 to adjust the model's weights during training.

This markdown content describes the structure and purpose of the `gausnet` model, highlighting its sequential architecture and the rationale behind the selection of layers and operations for classifying synthetic Gaussian blob images.


```python
# create a class for the model
def makeTheNet():

  class gausnet(nn.Module):
    def __init__(self):
      super().__init__()
      
      # all layers in one go using nn.Sequential
      self.enc = nn.Sequential(
          nn.Conv2d(1,6,3,padding=1),  # output size: (91+2*1-3)/1 + 1 = 91
          nn.ReLU(),                   # note that relu is treated like a "layer"
          nn.AvgPool2d(2,2),           # output size: 91/2 = 45 
          nn.Conv2d(6,4,3,padding=1),  # output size: (45+2*1-3)/1 + 1 = 45
          nn.ReLU(),                   # 
          nn.AvgPool2d(2,2),           # output size: 45/2 = 22
          nn.Flatten(),                # vectorize conv output
          nn.Linear(22*22*4,50),       # output size: 50
          nn.Linear(50,1),             # output size: 1
      )
      
    def forward(self,x):
      return self.enc(x)
  
  # create the model instance
  net = gausnet()
  
  # loss function
  lossfun = nn.BCEWithLogitsLoss()

  # optimizer
  optimizer = torch.optim.Adam(net.parameters(),lr=.001)

  return net,lossfun,optimizer
```

```python
# test the model with one batch
net,lossfun,optimizer = makeTheNet()

X,y = next(iter(train_loader))
yHat = net(X)

# check size of output
print(' ')
print(yHat.shape)

# # now let's compute the loss
loss = lossfun(yHat,y)
print(' ')
print('Loss:')
print(loss)
```

```python
# count the total number of parameters in the model
summary(net,(1,imgSize,imgSize))
```

# Create a function that trains the model

```python
# a function that trains the model

def function2trainTheModel():

  # number of epochs
  numepochs = 10
  
  # create a new model
  net,lossfun,optimizer = makeTheNet()

  # initialize losses
  trainLoss = torch.zeros(numepochs)
  testLoss  = torch.zeros(numepochs)
  trainAcc  = torch.zeros(numepochs)
  testAcc   = torch.zeros(numepochs)


  # loop over epochs
  for epochi in range(numepochs):

    # loop over training data batches
    batchLoss = []
    batchAcc  = []
    for X,y in train_loader:

      # forward pass and loss
      yHat = net(X)
      loss = lossfun(yHat,y)

      # backprop
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

      # loss from this batch
      batchLoss.append(loss.item())
      batchAcc.append( torch.mean(( (yHat>0) ==y).float()).item() )
    # end of batch loop...

    # and get average losses across the batches
    trainLoss[epochi] = np.mean(batchLoss)
    trainAcc[epochi]  = 100*np.mean(batchAcc)

    # test accuracy
    X,y = next(iter(test_loader)) # extract X,y from test dataloader
    with torch.no_grad(): # deactivates autograd
      yHat = net(X)
      loss = lossfun(yHat,y)
      
    # compare the following really long line of code to the training accuracy lines
    testLoss[epochi] = loss.item()
    testAcc[epochi]  = 100*torch.mean(( (yHat>0) ==y).float()).item()

  # end epochs

  # function output
  return trainLoss,testLoss,trainAcc,testAcc,net
```

# Run the model and show the results!

```python
trainLoss,testLoss,trainAcc,testAcc,net = function2trainTheModel()
```

```python
fig,ax = plt.subplots(1,2,figsize=(16,5))

ax[0].plot(trainLoss,'s-',label='Train')
ax[0].plot(testLoss,'o-',label='Test')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss (MSE)')
ax[0].set_title('Model loss')

ax[1].plot(trainAcc,'s-',label='Train')
ax[1].plot(testAcc,'o-',label='Test')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy (%)')
ax[1].set_title(f'Final model test accuracy: {testAcc[-1]:.2f}%')
ax[1].legend()

plt.show()
```
